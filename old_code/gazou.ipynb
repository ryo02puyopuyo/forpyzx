```python
import pyzx as zx


# PyZXで量子回路を作成
circuit = zx.Circuit(2)  # 量子ビット数2

# ゲートの追加 #kairo
circuit.add_gate("T", 0)        # 量子ビット0にアダマールゲート
circuit.add_gate("S",1)
circuit.add_gate("CNOT", 0, 1)  # 量子ビット0と1に制御NOTゲート
circuit.add_gate("CNOT", 1, 0)
circuit.add_gate("S",1)
circuit.add_gate("H",0)



# 回路を可視化
zx.draw(circuit)
```

```python
from qiskit import QuantumCircuit
from qiskit.visualization import circuit_drawer

# 量子回路の作成
qc = QuantumCircuit(2)  # 2量子ビットの回路

# kairo と同じ回路を作成
qc.t(0)  # 量子ビット0にアダマールゲート
qc.s(1)
qc.cx(0, 1)  # 量子ビット0と1に制御NOTゲート
qc.cx(1, 0)
qc.s(1)
qc.h(0)
qc.tdg(1)


# 回路を描画して表示
qc.draw('mpl')  # matplotlibを用いて描画

# PNG画像として保存
circuit_drawer(qc, output='mpl', filename='circuit.png')    
```
