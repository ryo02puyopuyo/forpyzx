```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import openpyxl
```

```python
df = pd.read_excel("/home/rest/forpyzx/excel/202412015outputchukan.xlsx",sheet_name='Sheet1',header=1)
```

```python
qubits = [4,16]
gates = 50000
p_t_lists = [i * 0.04 for i in range(0, 11)]
optimize_name = ["original","basic","clifford_simp"]

#fig,axes = plt.subplots(7,11)
fig, ax = plt.subplots(len(qubits),figsize=(7.5, 10),constrained_layout=True)

#p_t
for qubit in range(len(qubits)):
        ax[qubit].set_title(f"{qubits[qubit]} qubits, {gates} gates")
        ax[qubit].set_xlabel("p_t")
        ax[qubit].set_ylabel("all gates")
        ax[qubit].set_xlim(min(p_t_lists),max(p_t_lists)+0.02)
        ax[qubit].set_xticks(p_t_lists)
        ax[qubit].set_ylim(0,gates*1.25)
        ax[qubit].grid(True)

        for optimize in optimize_name:
            data = df[(df["optimize_name"] == optimize) & (df["gates_bef"] == gates) & (df["qubits"] == qubits[qubit]) ]
            if optimize == "clifford_simp" :
                plot_style = 'x'
            elif optimize == "basic":
                plot_style = 'v'
            else: 
                plot_style = 'o'

            #data2 = data[data["p_t"] == 0.04]
            #print(data2, optimize)
            
            
            mean_data = (data.groupby("p_t", as_index=False)["gates_aft"].mean())
            #if optimize == "original":
            #    print(mean_data[mean_data["p_t"] == 0.04] , optimize)
            print(mean_data[mean_data["p_t"] == 0.04] , optimize ,qubits[qubit])

            ax[qubit].plot(mean_data["p_t"],mean_data["gates_aft"],plot_style,linestyle='-',markersize=7,label=optimize)
            #ax[qubit].legend()

plt.show()
```
