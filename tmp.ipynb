```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib


# ---  ---
# 
threads = np.array([1, 2, 4, 8, 16])

#  ()
times = np.array([0.840310, 0.421148, 0.217244, 0.117375, 0.062017])


# ---  ---
# 1
baseline_time = times[0] 
#  = (1) / (N)
speedup_rates = baseline_time / times


# ---  ---
# 
# Windows: 'Yu Gothic', 'Meiryo'
# Mac: 'Hiragino Sans'
# Linux: 'IPAexGothic' (: sudo apt-get install fonts-ipaexfont)
plt.rcParams['font.family'] = 'IPAexGothic'

# 
plt.figure(figsize=(8, 6))

# 1.  ()
plt.plot(threads, speedup_rates, marker='o', linestyle='-', color='blue', label='')

# 2.  ()
#  (: 44)
plt.plot(threads, threads, linestyle='--', color='red', label=' (N)')

# ---  ---
# 
plt.title('', fontsize=16)

# 
plt.xlabel('', fontsize=12)
plt.ylabel(' ()', fontsize=12)

# X
plt.xticks(threads)

# 
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# 
plt.legend(fontsize=12)

# 
plt.show()

# ---  ---
print("---  ---")
for i in range(len(threads)):
    print(f": {threads[i]}, : {speedup_rates[i]:.4f}")


plt.savefig("/home/rest/forpyzx/speedup.pdf", format="pdf")
```
